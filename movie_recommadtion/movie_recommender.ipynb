```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(10)
```

```python
movies.head(1)['keywords'].values
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python


def casting(obj):
    C =[]
    counter = 0 
    for i in ast.literal_eval(obj):
        if counter != 3:
            C.append(i['name'])
            counter +=1
        else:
             break
    return C
```

```python
 movies['cast'] = movies['cast'].apply(casting)
```

```python
def fetch_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
 movies['crew'] = movies['crew'].apply(fetch_crew)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tag'] = movies['overview'] +movies['genres']+movies['keywords']+movies['cast'] + movies['crew']
```

movies.head(1)

```python
movies.head(1)
```

```python
df = movies[['movie_id','title','tag']]
```

```python
df.head(1)
```

```python
df['tag'] = df['tag'].apply(lambda x : " ".join(x))
```

```python
df.head()
```

```python
df['tag'] = df['tag'].apply(lambda x : x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tag']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tag'] = df['tag'].apply(stem)
```

```python
df.head(1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate( similarity[119])),reverse=True ,key=lambda x:x[1])[1:6]
```

```python
df[df['title'] == 'Batman Begins'].index[0]
```

```python
def recommed(text):
    movie_index = df[df['title'] == text].index[0]
    distance  = similarity[movie_index]
    movie_list = sorted(list(enumerate( distance)),reverse=True ,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommed('Spectre')
```
